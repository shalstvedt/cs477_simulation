```python
from __future__ import divisionimport numpyfrom scipy.integrate import odeintimport pylab
```

## Constants

```python
dt = 0.01Tend = 200.0
```

## Helper functions

```python
# Unpacks 2-vectors of the form [x1,y1,vx1,vy1,x2,y2,vx2,vy2, ...]def unpack(state):	ret = []	for i in numpy.arange(len(state)//2)*2:		ret.append(numpy.array([state[i], state[i+1]]))	return ret	## Calculates the length of 2-vectorsdef h(r):	return numpy.hypot(r[0],r[1])
```

## Equation of motion

```python
def electron_motion_two(state, t):	r1,v1,r2,v2 = unpack(state)	r21  = r2 - r1				a1 = -2*r1/(h(r1)**3)-r21/(h(r21)**3)	a2 = -2*r2/(h(r2)**3)+r21/(h(r21)**3)		return numpy.array([v1[0], v1[1], a1[0], a1[1], v2[0], v2[1], a2[0], a2[1]])
```

## Properties of the system### Angular momentum

```python
def angular_momentum_two(state):	r1,v1,r2,v2 = unpack(state)		l1 = r1[0]*v1[1]-r1[1]*v1[0]	l2 = r2[0]*v2[1]-r2[1]*v2[0]		return l1+l2
```

### Energy

```python
def energy_two(state):	r1,v1,r2,v2 = unpack(state)	r21  = r2 - r1		a1 = -2*r1/(h(r1)**3)+r21/(h(r21)**3)	a2 = -2*r2/(h(r2)**3)-r21/(h(r21)**3)		p1 = -a1*h(v1)*dt	p2 = -a2*h(v2)*dt		k1 = 0.5*h(v1)**2	k2 = 0.5*h(v2)**2		return k1+k2+h(p1)+h(p2)
```

## Plotting function

```python
def plot_two(odeint_result, title, xl, yl):	pylab.figure(figsize=(10,10))        pylab.title(title)	pylab.plot([e[0] for e in odeint_result], [e[1] for e in odeint_result])	pylab.plot([e[4] for e in odeint_result], [e[5] for e in odeint_result])	pylab.xlim(xl)	pylab.ylim(yl)		pylab.figure(figsize=(10,4))	pylab.plot([energy_two(e) for e in odeint_result])	pylab.title("Total energy of the Helium atom system over time")	pylab.xlabel('Time')	pylab.ylabel('Total Energy')		pylab.figure(figsize=(10,4))	pylab.plot([angular_momentum_two(e) for e in odeint_result])	pylab.title("Angular momentum of the Helium atom system over time")	pylab.xlabel('Time')	pylab.ylabel('Angular Momentum')
```

## Integration and initial conditions### Set up the time array

```python
times = numpy.linspace(0,Tend,int(Tend/dt))
```

### First initial condition

```python
c1 = numpy.array([2,0,0.95,0,-1,0,0,-1])res1 = odeint(electron_motion_two,c1,times)plot_two(res1,"First initial condition",(-2,2),(-2,2))
```

### Second initial condition

```python
c2 = numpy.array([1.4, 0.0, 0, 0.86, -1, 0.0, 0.0, -1])res2 = odeint(electron_motion_two,c2,times)plot_two(res2,"Second initial condition",(-2,2), (-2, 2))
```

### Third initial condition

```python
c3 = numpy.array([3, 0.0, 0, 0.4, 1, 0.0, 0.0, -1])res3 = odeint(electron_motion_two,c3,times)plot_two(res3,"Third initial condition",(-6,6), (-6, 6))
```

### Fourth initial condition

```python
c4 = array([2.5, 0.0, 0, 0.4, 1, 0.0, 0.0, -1])res4 = odeint(electron_motion_two,c4,times)plot_two(res4,"Fourth initial condition",(-1,3), (-2.5, 2.5))
```
