```python
def runge_kutta(f,t,x0,dt):    k1 = f(t,x0)*dt    k2 = f(t+dt/2,x0+k1/2)*dt    k3 = f(t+dt/2,x0+k2/2)*dt    k4 = f(t+dt,x0+k3)*dt    return x0+(k1+2*k2+2*k3+k4)/6
```

```python
def eom_odeint(state, t):    r1   = array([state[0], state[1]])    r2   = array([state[4], state[5]])    r3   = array([state[0]+2,state[1]+2])    r4   = array([state[4]+2,state[5]+2])       r21  = r2 - r1        v1   = array([state[2], state[3]])    v2   = array([state[6], state[7]])    hr1  = hypot(r1[0], r1[1])    hr2  = hypot(r2[0], r2[1])    hr3  = hypot(r3[0], r3[1])    hr4  = hypot(r4[0], r4[1])    hr21 = hypot(r21[0], r21[1])        a1 = -GM*r1/(hr1**3) + GM*m2*r21/(hr21**3)    a2 = -GM*r2/(hr2**3) - GM*m1*r21/(hr21**3)        return array([v1[0], v1[1], a1[0], a1[1],                  v2[0], v2[1], a2[0], a2[1]])
```

```python
def eom_2_odeint(state, t):    r1 = array([state[0], state[1]])    r2 = array([state[0]-2,state[1]])        v = array([state[2], state[3]])        hr1 = hypot(r1[0], r1[1])    hr2 = hypot(r2[0], r2[1])        a = -GM*r1/(hr1**3) -GM*r2/(hr2**3)        return array([v[0], v[1], a[0], a[1]])
```

```python
def accel(planet, otherplanet, mass):    r1 = (planet[POS1][X]**2 + planet[POS1][Y]**2)**0.5    r2 = ((otherplanet[POS1][X]-planet[POS1][X])**2+(otherplanet[POS1][Y]-planet[POS1][Y])**2)**0.5    eq = ((-GM*planet[POS1])/r1**3)-(GM*mass*(planet[POS1]-otherplanet[POS1])/r2**3)    return array([planet[VEL1], eq])
```

```python
def accel_odeint(planet, otherplanet, mass):    r1 = array([planet[0], planet[1]])    r2 = array([otherplanet[0], otherplanet[1]])    r21 = r2 - r1    hr1 = hypot(r1[0], r1[1])    hr2 = hypot(r2[0], r2[1])    hr21 = hypot(r21[0], r21[1])    eq = (-GM*r1)/(hr1**3)+(GM*mass*r21)/(hr21**3)    if mass == m2:        eq = (-GM*r1)/(hr1**3)+(GM*mass*r21)/(hr21**3)    return array([planet[2:4], eq])
```

```python
def energy_odeint(y):    r1   = array([y[0], y[1]])    v1   = array([y[2], y[3]])    r2   = array([y[4], y[5]])    v2   = array([y[6], y[7]])    r21  = r2 - r1    hr1  = hypot(r1[0], r1[1])    hv1  = hypot(v1[0], v1[1])    hr2  = hypot(r2[0], r2[1])    hv2  = hypot(v2[0], v2[1])    hr21 = hypot(r21[0], r21[1])        return 0.5*m1*hv1**2+0.5*m2*hv2**2-GM*m1/hr1-GM*m2/hr2-GM*m2*m1/hr21
```

```python
def energy(y):    return energy_odeint(numpy.append(y[0], y[1]))
```

```python
def momentum_odeint(state):    x1, y1, vx1, vy1, x2, y2, vx2, vy2 = state    return m1*(x1*vy1-y1*vx1) + m2*(x2*vy2-y2*vx2)
```

```python
def momentum(y):    return momentum_odeint(numpy.append(y[0], y[1]))
```
